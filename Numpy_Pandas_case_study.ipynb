#Pandas Answers

##1. Basic Exploration

```python
import pandas as pd

df = pd.read_csv('/content/auto-mpg.csv')

df.head(10)

df.shape

df.describe()
```

##2. Filtering and Indexing

```python
df[(df['model year'] == 75) & (df['weight'] < 3000)][['car name', 'weight', 'mpg']]
```

##3. Handling Missing Data

```python
df.isna().sum()
```

```python
#converting horsepower to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
# Fill missing values with median
median_hp = df['horsepower'].median()
df.fillna({'horsepower': median_hp}, inplace=True)

df['horsepower'].head()



```

##4. Data Transformation

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df['power_to_weight_ratio']
```

##5. Group By

```python
origin_groups = df.groupby('origin')
origin_groups['mpg'].mean()
```

##6. Sorting

```python
df.sort_values(by='mpg', ascending=False).head(10)
```

##Apply function

###7. Apply this function to each row and store the result in the new column

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
df['performance_score']

```

###8. Visualization Preparation

```python
df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
```

###9. Exporting Data

```python
df[df['mpg'] > 30][['mpg','cylinders','horsepower','weight']].to_csv('high_mpg_cars.csv', index=False)
```

###10. Finding Anomalies (IQR)

```python
#Calculate the IQR for mpg
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values less than
#Q1 - 1.5 * IQR or greater than Q3 +1.5 * IQR.

outliers = df[(df['mpg'] < (Q1 - 1.5 * IQR)) | (df['mpg'] > (Q3 + 1.5 * IQR))]

#Create a DataFrame of cars classified as outliers,
#displaying car_name, mpg,and model_year.

outliers[['car name', 'mpg', 'model year']]
```

#Numpy Answers

```python
import numpy as np
```

###1. Basic Array Operations

```python
mpg = df['mpg'].values

np.mean(mpg)
np.median(mpg)
np.std(mpg)

np.sum(mpg > 25)

print(f"Mean: {np.mean(mpg):.2f}")
print(f"Median: {np.median(mpg):.2f}")
print(f"Std Dev: {np.std(mpg):.2f}")
print(f"Count > 25 mpg: {np.sum(mpg > 25)}")

```

###2. Filtering

```python
cars = df['car name'].values
cyl = df['cylinders'].values

cars[cyl > 6].tolist()
```

###3. Percentiles

```python
weight = df['weight'].values

np.percentile(weight, [25, 50, 75])
```

###4.Array Manipulation



```python
acc = df['acceleration'].values

# normalization (min-max scaling)
norm_acc = (acc - acc.min()) / (acc.max() - acc.min())

print(norm_acc)
```

###5. Broadcasting

```python
hp = df['horsepower'].values

# replace NaN
hp = np.where(np.isnan(hp), np.nanmean(hp), hp)

hp_new = hp * 1.1
print(hp_new[:10])
```

###6. Boolean Indexing


```python
# Convert columns to NumPy arrays
origin = df['origin'].values
displacement = df['displacement'].values

# Select only cars from Europe (origin == 2)
europe_disp = displacement[origin == 2]

# Calculate average
avg_disp = np.mean(europe_disp)

print(avg_disp)
```

###7. Matrix Operations

```python
matrix = df[['mpg','horsepower','weight']].values
vector = np.array([1, 0.5, -0.2])

np.dot(matrix, vector)

```

###8. Sorting

```python
sorted_idx = np.argsort(df['model year'].values)[::-1]

df['car name'].values[sorted_idx][:5]
```

###9. Correlation

```python
mpg = df['mpg'].values
weight = df['weight'].values

np.corrcoef(mpg, weight)
```

###10. Conditional Aggregates

```python
cyl = df['cylinders'].values
mpg = df['mpg'].values

for c in np.unique(cyl):
    print(c, np.mean(mpg[cyl == c]))
```
